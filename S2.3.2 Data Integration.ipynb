## Importing the Filtered Datasets

```python
# Code Contributed by - Asif Rana   StudentID- 102866893

# Importing pandas library
import pandas as pd
import numpy as np
import random
# Importing the csv data files 
data_links = pd.read_csv('datasets_links_small.csv',error_bad_lines=False, warn_bad_lines=True)
ratings = pd.read_csv('ratings_filtered.csv',error_bad_lines=False, warn_bad_lines=True)
movies_metadata = pd.read_csv('metadata_filtered.csv',error_bad_lines=False, warn_bad_lines=True, low_memory=False)
```

## Integrating tmdbId and imdbId to ratings_small.csv 

```python
# Using the inner join to integrate imdbId and tmdbId to the ratings dataframe
ratings_movielink_df = pd.merge(ratings, data_links, how='inner', left_on=['movieId'], right_on = ['movieId'])
```

## Integrating movieId to movies_metadata.csv using link_small

```python
# Using the inner join to integrate movieId to the ratings dataframe
# Using inner join will remove the movies which are not present in the rating.csv file

movies_metadata_integrated = pd.merge(movies_metadata, data_links, how='inner', left_on=['tmdbId'], right_on = ['tmdbId']).drop(columns = ['imdbId'])
```
