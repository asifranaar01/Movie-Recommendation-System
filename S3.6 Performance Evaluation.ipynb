```python
# Hyperparameter Tuning for Collaborative Filter (model-based)
```

```python
#plotting rmse scores for each k for memory-based user-based collaborative filter
k = [5,50,100,150,200]
RMSE = [0.9506, 0.9921, 0.9435, 0.9195, 0.9120]
plt.plot(k, RMSE)
plt.title("RMSE for each k")
plt.xlabel('k')
plt.ylabel('RMSE')
plt.show()
```

As k increases, the RMSE score decreases. However, the computational time increases with every increase in k. Most importantly, increasing k causes overfitting upto a point where, when k = 100, the rmse score for training data falls down to 0.0627. Hence, the best acceptable value for k is 5.
